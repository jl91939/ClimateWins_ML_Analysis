# Importing Libraries + Setting Path.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from numpy import reshape
from keras.models import Sequential
from keras.layers import Conv1D, Conv2D, Dense, BatchNormalization, Flatten, MaxPooling1D, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import LSTM

path = r'C:\Users\Chaot\Python stuff\07 - Basics of Machine Learning for Analysts'
```

# Importing Data.

```python
unscaled = pd.read_csv(os.path.join(path, 'Data Brief + Data Sets', 'Dataset-weather-prediction-dataset-processed.csv'))
unscaled
```

```python
pleasant = pd.read_csv(os.path.join(path, 'Data Brief + Data Sets', 'Dataset-Answers-Weather_Prediction_Pleasant_Weather.csv'))
pleasant
```

# Cleaning the Observations & Predictions data frame.

```python
unscaled = unscaled.drop(['DATE','MONTH',
'GDANSK_cloud_cover',
'GDANSK_humidity',
'GDANSK_precipitation',
'GDANSK_snow_depth',
'GDANSK_temp_mean',
'GDANSK_temp_min',
'GDANSK_temp_max',
'ROMA_cloud_cover',
'ROMA_wind_speed',
'ROMA_humidity',
'ROMA_pressure',
'ROMA_sunshine',
'ROMA_temp_mean',
'TOURS_wind_speed',
'TOURS_humidity',
'TOURS_pressure',
'TOURS_global_radiation',
'TOURS_precipitation',
'TOURS_temp_mean',
'TOURS_temp_min',
'TOURS_temp_max'], axis=1)
unscaled
```

```python
pleasant = pleasant.drop(['DATE'], axis=1)
pleasant.shape
```

```python
columns_to_drop = ('_snow_depth', '_wind_speed')
unscaled = unscaled.loc[:, ~unscaled.columns.str.endswith(columns_to_drop)]
unscaled
```

```python
position1 = unscaled.columns.get_loc('KASSEL_humidity')-1
unscaled.insert(position1,'KASSEL_cloud_cover',unscaled['LJUBLJANA_cloud_cover'])

position2 = unscaled.columns.get_loc('MUNCHENB_global_radiation')-1
unscaled.insert(position2,'MUNCHENB_pressure',unscaled['SONNBLICK_pressure'])

position3 = unscaled.columns.get_loc('STOCKHOLM_pressure')-1
unscaled.insert(position3,'STOCKHOLM_humidity',unscaled['OSLO_humidity'])

unscaled.shape
```

```python
unscaled.to_csv(os.path.join(path, 'Data Brief + Data Sets', 'cleaned_weather.csv'), index=False)
```

# Creating Array and Reshaping X.

```python
X = unscaled
y = pleasant
X
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X = X.reshape(-1,15,9)
X.shape
```

```python
y.shape
```

# Creating Test/Train sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
predictions = {0: 'BASEL', 1: 'BELGRADE', 2: 'BUDAPEST', 3: 'DEBILT', 4: 'DUSSELDORF', 5: 'HEATHROW', 6: 'KASSEL', 7: 'LJUBLJANA',
8: 'MAASTRICHT', 9: 'MADRID', 10: 'MUNCHENB', 11: 'OSLO', 12: 'SONNBLICK', 13: 'STOCKHOLM', 14: 'VALENTIA'}
```

# RNN Model.

```python
epochs = 5
batch_size = 8
n_hidden = 16

timesteps = len(X_train[0])
input_dim = len(X_train[0][0])
n_classes = len(y_train[0])

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='sigmoid')) #Don't use relu here!
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
model.fit(X_train,
          y_train,
          batch_size=batch_size,
          validation_data=(X_test, y_test),
          epochs=epochs)
```

```python
epochs = 5
batch_size = 8
n_hidden = 16

timesteps = len(X_train[0])
input_dim = len(X_train[0][0])
n_classes = len(y_train[0])

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='tanh')) #Don't use relu here!
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
model.fit(X_train,
          y_train,
          batch_size=batch_size,
          validation_data=(X_test, y_test),
          epochs=epochs)
```

```python
y_train.shape
```

```python
def confusion_matrix(Y_true, Y_pred):
    Y_true = pd.Series([predictions[y] for y in np.argmax(Y_true, axis=-1)])
    Y_pred = pd.Series([predictions[y] for y in np.argmax(Y_pred, axis=-1)])

    return pd.crosstab(Y_true, Y_pred, rownames=['True'], colnames=['Pred'])

pred = model.predict(X_test)
pred.shape
```

```python
y_test.shape
```

```python
# Evaluate
print(confusion_matrix(y_test, pred))
```
