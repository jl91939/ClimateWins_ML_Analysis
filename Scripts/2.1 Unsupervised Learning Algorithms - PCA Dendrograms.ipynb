# Importing Libraries + Setting Path.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from matplotlib.pyplot import figure
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist

path = r'C:\Users\Chaot\Python stuff\07 - Basics of Machine Learning for Analysts'
```

# Importing Dataframes.

```python
pca_df = pd.read_csv(os.path.join(path, 'Data Brief + Data Sets', 'PCA_df.csv'))
pca_df
```

# Dendrogram - Single.

```python
distance_single = linkage(pca_df,method='single')
plt.figure(figsize=(20,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle('Dendrogram Single Method - PCA',fontsize=18)
plt.savefig(os.path.join(path, 'Unsupervised', '2.1_dendrogram_single_pca.png'), bbox_inches='tight')
plt.show()
```

# Dendrogram - Complete.

```python
distance_single = linkage(pca_df,method='complete')
plt.figure(figsize=(20,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle('Dendrogram Complete Method - PCA',fontsize=18)
plt.savefig(os.path.join(path, 'Unsupervised', '2.1_dendrogram_complete_pca.png'), bbox_inches='tight')
plt.show()
```

# Dendrogram - Average.

```python
distance_single = linkage(pca_df,method='average')
plt.figure(figsize=(20,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle('Dendrogram Average Method - PCA',fontsize=18)
plt.savefig(os.path.join(path, 'Unsupervised', '2.1_dendrogram_average_pca.png'), bbox_inches='tight')
plt.show()
```

# Dendrogram - Ward.

```python
distance_single = linkage(pca_df,method='ward')
plt.figure(figsize=(20,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle('Dendrogram Ward Method - PCA',fontsize=18)
plt.savefig(os.path.join(path, 'Unsupervised', '2.1_dendrogram_ward_pca.png'), bbox_inches='tight')
plt.show()
```
